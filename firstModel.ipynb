```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [173, 171, 189, 181]
y = [81, 72, 96, 94]

x_ones = np.c_[np.ones(4), x]
x_ones
```

```python
theta = np.linalg.inv(x_ones.T.dot(x_ones)).dot(x_ones.T).dot(y)
theta
```

```python
plt.scatter(x,y,s=80,c = '#06d6a0')

x_lim = [170,190]
x_lim_ones = np.c_[np.ones(2), x_lim]
y_lim = x_lim_ones.dot(theta)
plt.plot(x_lim, y_lim, 'b-')

plt.axis([170, 190, 70,100])
plt.xlabel('Altura')
plt.ylabel ('Peso')
plt.title('Altura x peso')
plt.grid()
```
